```python
import itertools

import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
from scipy import linalg

from sklearn import mixture

import pandas as pd

import random

from sklearn.mixture import BayesianGaussianMixture
```

```python
V = pd.read_csv("~/Desktop/VI submission/code/outputs/V_GW1722.csv", index_col=0)
```

```python
V
```

```python
K = [14]
weightP = [1/14]
```

```python
ddgmmLabel=[]
ddgmm_prob = []
for k in K:
  for w in weightP:
    dpgmm = mixture.BayesianGaussianMixture(
        n_components = k,
        covariance_type="full",
        weight_concentration_prior= w,
        weight_concentration_prior_type="dirichlet_distribution",
        mean_precision_prior=1e-2,
        covariance_prior=1e0 * np.eye(16),
        init_params="kmeans",
        max_iter=200,
        random_state=2,
    ).fit(V)

  ddgmmLabel.append(dpgmm.predict(V))
  ddgmm_prob.append(dpgmm.predict_proba(V))
```
